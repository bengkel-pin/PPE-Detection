```python
project = rf.workspace("kelvin-snkig").project("ppe-akpdb")

# Path ke file train/valid/test
file_path = "./CHV_dataset/data split/train.txt"

with open(file_path, "r") as file:
    image_paths = [line.strip() for line in file if line.strip()]

for i, image_path in enumerate(image_paths):
    # Pastikan file tidak diawali "._" dan benar-benar file gambar
    filename = os.path.basename(image_path)
    if (
        not filename.startswith("._") and
        filename.lower().endswith((".jpg", ".jpeg", ".png", ".bmp", ".webp"))
    ):
        try:
            project.upload(
                image_path=image_path,
                batch_name="batch_train",
                split="train",
                num_retry_uploads=3,
                tag_names=[],
                sequence_number=i,
                sequence_size=len(image_paths)
            )
            print(f"Uploaded: {image_path} ({i+1}/{len(image_paths)})")
        except Exception as e:
            print(f"Failed to upload {image_path}: {e}")
```
